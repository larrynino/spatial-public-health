```python
# Función para calcular porcentajes de cada categoría
def calcular_porcentajes(grupo, columna, valores_map):
    """
    Calcula el porcentaje de cada valor en la columna para el grupo
    """
    total = len(grupo)
    porcentajes = {}
    
    for valor, nombre_col in valores_map.items():
        count = (grupo[columna] == valor).sum()
        porcentajes[nombre_col] = (count / total * 100) if total > 0 else 0
    
    return pd.Series(porcentajes)

# Definir mapeos de valores a nombres de columnas
valores_pared = {
    1: 'PA_1FIR',
    2: 'PA_2MBU',
    3: 'PA_3MVE',
    4: 'PA_4BRR',
    5: 'PA_5MRE',
    6: 'PA_6NOP'
}

valores_piso = {
    1: 'PI_1FIR',
    2: 'PI_2MBU',
    3: 'PI_3DES'
}

valores_sersa = {
    1: 'IN_1CON',
    2: 'IN_2NOC',
    3: 'IN_3DES',
    4: 'IN_4NOI'
}

# Calcular porcentajes por manzana para cada columna (con include_groups=False)
porcentajes_pared = df_viv_mzn_nan_reclas.groupby('COD_DANE_ANM').apply(
    lambda x: calcular_porcentajes(x, 'V_MAT_PARED', valores_pared),
    include_groups=False
).reset_index()

porcentajes_piso = df_viv_mzn_nan_reclas.groupby('COD_DANE_ANM').apply(
    lambda x: calcular_porcentajes(x, 'V_MAT_PISO', valores_piso),
    include_groups=False
).reset_index()

porcentajes_sersa = df_viv_mzn_nan_reclas.groupby('COD_DANE_ANM').apply(
    lambda x: calcular_porcentajes(x, 'V_TIPO_SERSA', valores_sersa),
    include_groups=False
).reset_index()

# Combinar todos los porcentajes en un solo dataframe
df_viv_mzn_nan_reclas_agg = porcentajes_pared.merge(
    porcentajes_piso, on='COD_DANE_ANM'
).merge(
    porcentajes_sersa, on='COD_DANE_ANM'
)

# Verificar resultados
print(f"Total de manzanas: {len(df_viv_mzn_nan_reclas_agg)}")
print(f"\nColumnas generadas: {df_viv_mzn_nan_reclas_agg.columns.tolist()}")
print(f"\nPrimeras filas:")
print(df_viv_mzn_nan_reclas_agg.head())
```

```python
# Función para calcular porcentajes de cada categoría
def calcular_porcentajes(grupo, columna, valores_map):
    """
    Calcula el porcentaje de cada valor en la columna para el grupo
    """
    total = len(grupo)
    porcentajes = {}
    
    for valor, nombre_col in valores_map.items():
        count = (grupo[columna] == valor).sum()
        porcentajes[nombre_col] = (count / total * 100) if total > 0 else 0
    
    return pd.Series(porcentajes)

# Crear una copia del dataframe para no modificar el original
df_temp = df_hog_mzn_nan_reclas.copy()

# Limpiar valores a ignorar (convertir a NaN)
df_temp['H_NRO_CUARTOS'] = df_temp['H_NRO_CUARTOS'].replace(99, pd.NA)
df_temp['H_NRO_DORMIT'] = df_temp['H_NRO_DORMIT'].replace(99, pd.NA)
df_temp['H_DONDE_PREPALIM'] = df_temp['H_DONDE_PREPALIM'].replace(9, pd.NA)
df_temp['H_AGUA_COCIN'] = df_temp['H_AGUA_COCIN'].replace(99, pd.NA)

# Definir mapeos de valores a nombres de columnas
valores_cocina = {
    1: 'CO_1EXC',
    2: 'CO_2COM',
    3: 'CO_3PER',
    4: 'CO_4NOC'
}

valores_agua = {
    1: 'AG_1ENO',
    2: 'AG_2ACU',
    3: 'AG_3POZ',
    4: 'AG_4COM',
    5: 'AG_5HID',
    6: 'AG_6LLU'
}

# Calcular sumas de cuartos y dormitorios por manzana (ignorando NaN)
sumas_cuartos_dormit = df_temp.groupby('COD_DANE_ANM').agg({
    'H_NRO_CUARTOS': 'sum',
    'H_NRO_DORMIT': 'sum'
}).rename(columns={
    'H_NRO_CUARTOS': 'NCUA',
    'H_NRO_DORMIT': 'NDOR'
}).reset_index()

# Calcular porcentajes por manzana para cada columna (con include_groups=False)
porcentajes_cocina = df_temp.groupby('COD_DANE_ANM').apply(
    lambda x: calcular_porcentajes(x, 'H_DONDE_PREPALIM', valores_cocina),
    include_groups=False
).reset_index()

porcentajes_agua = df_temp.groupby('COD_DANE_ANM').apply(
    lambda x: calcular_porcentajes(x, 'H_AGUA_COCIN', valores_agua),
    include_groups=False
).reset_index()

# Combinar todos los dataframes
df_hog_mzn_nan_reclas_agg = porcentajes_cocina.merge(
    porcentajes_agua, on='COD_DANE_ANM'
).merge(
    sumas_cuartos_dormit, on='COD_DANE_ANM'
)

# Verificar resultados
print(f"\nTotal de manzanas: {len(df_hog_mzn_nan_reclas_agg)}")
print(f"\nColumnas generadas: {df_hog_mzn_nan_reclas_agg.columns.tolist()}")
print(f"\nPrimeras filas (todas las columnas):")
print(df_hog_mzn_nan_reclas_agg.head())
```
